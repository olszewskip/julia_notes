# Exploring the Julia Set in Julia

```julia
const maxiter = 100

"""Compute the "escape time" of the Julia set at (z,c)."""

function julia(z, c)
    for n = 1:maxiter
        abs2(z) > 4 && return n - 1
        z = z * z + c
    end
    return maxiter
end
```

```julia
julia(0.4 + 0.5im, 0.6 + 0.7im)
```

```julia
[julia(r + i*im, -0.06 + 0.665im) for i=-0.5:0.002:0.5, r=-1:0.002:1]
```

```julia
using Colors, Images, ImageMagick
```

```julia
RGB{FixedPointNumbers.N0f8}
```

```julia
const cmap = convert(Vector{RGB{FixedPointNumbers.N0f8}}, colormap("RdBu", 100))
```

```julia
[cmap[julia(r + i*im, -0.04 + 0.7*im)] for i = -1:0.005:1, r=-1.5:0.005:1.5]
```

```julia
[cmap[julia(r + i*im, r + i*im)] for i = -1:0.005:1, r=-1.5:0.005:1.5]
```

```julia
using Interact
```

```julia
@manipulate for r_ = -0.1 : 0.01 : 0.1, i_ = 0.65 : 0.01 : 0.85
    [cmap[julia(r + i*im, r_ + i_*im)] for i=-1:0.005:1, r=-1.5:0.005:1.5]
end
```

```julia
let I = -1:0.005:1, R = -1:0.005:1.5
    data = Array{RGB{FixedPointNumbers.N0f8}}(undef,(length(I), length(R)))
    @manipulate for r_ = -0.1 : 0.005 : 0.1, i_ = 0.65 : 0.005 : 0.85
        for (j, i) in enumerate(I),
            (k, r) in enumerate(R)
            data[j, k] = cmap[julia(r + i*im, r_ + i_*im)]
        end
        data
    end
end
```
